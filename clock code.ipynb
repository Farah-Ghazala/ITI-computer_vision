```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from math import cos,sin,radians
import math
import datetime

cv.namedWindow("Clock")
img = np.zeros((500,500,3), np.uint8)
cv.circle(img,(250,250),200,[255,255,255],3)
cv.circle(img,(250,250),4,[0, 255, 255],-1)

center=(250,250)
radious=200

def get_tick():
    start_point=[]
    end_point=[]
    
    for i in range(0,360,6):   
        x=int((center[0]+radious*cos(radians(i))))
        y=int((center[1]+radious*sin(radians(i))))
        start_point.append((x,y))
        
    for i in range(0,360,6):   
        x=int((center[0]+(radious-20)*cos(radians(i))))
        y=int((center[1]+(radious-20)*sin(radians(i))))
        end_point.append((x,y))

    return start_point,end_point



def draw_clock(img):
    #apply get_tick function
    start_point,end_point = get_tick()
    for i in range(len(start_point)):
        if i % 5 == 0:
            cv.line(img, start_point[i], end_point[i], (0, 255, 255), 3)
            #cv.circle(img, start_point[i], 4, (0, 255, 0), -1)
        
        else:
            cv.circle(img, start_point[i], 2, (255, 0, 0), -1)
     
    #draw clock hours
    for i in range(1, 13):
        angle = math.radians(i * 30 + 270)
        number_x = int((center[0]-5) + (radious + 15) * math.cos(angle))
        number_y = int((center[1]+5) + (radious + 15) * math.sin(angle))
        cv.putText(img, str(i), (number_x, number_y), cv.FONT_HERSHEY_SIMPLEX, 0.5 , (0, 255, 255), 2, cv.LINE_AA)

     #get time   
    now = datetime.datetime.now()
    hour = now.hour % 12
    minute = now.minute
    second = now.second
    microsecond = now.microsecond

    # Calculate angles of Clockwise
    hour_angle = math.radians((hour * 30) + (minute * 0.5) + 270)
    minute_angle = math.radians((minute * 6) + 270)
    second_angle = math.radians((second * 6) + (microsecond * (6 / 1000000)) + 270)

    # Calculate the hands
    hour_x = int(center[0] + (radious - 80) * math.cos(hour_angle))
    hour_y = int(center[1] + (radious - 80) * math.sin(hour_angle))
    minute_x = int(center[0] + (radious - 50) * math.cos(minute_angle))
    minute_y = int(center[1] + (radious - 50) * math.sin(minute_angle))
    second_x = int(center[0] + (radious - 25) * math.cos(second_angle))
    second_y = int(center[1] + (radious - 25) * math.sin(second_angle))
    
    # Draw hands
    cv.line(img, center, (hour_x, hour_y), (255, 255, 255), 4)
    cv.line(img, center, (minute_x, minute_y), (255, 255, 255), 2)
    cv.line(img, center, (second_x, second_y), (255, 255, 255), 1)
    
    return img
    
    
while True:
    my_image = img.copy()
    clock = draw_clock(my_image)
    
    now = datetime.datetime.now()
    date = now.strftime("%b %d, %Y")
    current_time = now.strftime("%H:%M:%S")

    # show the digital time
    cv.putText(clock, ('Date : '+date), (0, 20), cv.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2, cv.LINE_AA)
    cv.putText(clock, ('Time : '+current_time), (0, 46), cv.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2, cv.LINE_AA)

    
    cv.imshow("Clock",clock)
    if cv.waitKey(1) & 0xFF == ord('q'):
        break
cv.destroyAllWindows()
```
