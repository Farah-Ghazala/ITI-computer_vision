```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

### Q1 load the following image and make red rectangle around face and write below rectangle “Cat” with yellow color.

```python
img=cv2.imread("original.png")
cv2.rectangle(img,(780,120),(1550,830),(255,0,0),3)
cv2.putText(img,"Cat",(900,1200),cv2.FONT_HERSHEY_COMPLEX,10,(255,255,0),4)
plt.imshow(img)
```

### Q2 load the following image and apply sharpen filter.
### “write your observation”

### my observation: after applying the kernel the image become clearer

```python
img = cv2.imread('blurred.png')

kernel = np.array([
    [-1,-1,-1],
    [-1,9,-1],
    [-1,-1,-1]
])

sharpen_image = cv2.filter2D(img,-1,kernel)

f,(ax0,ax1)=plt.subplots(1,2,figsize=(30,15))

ax0.imshow(img)
ax1.imshow(sharpen_image)
```

### Q2 load the following image and apply blur filter 
### “write your observation”
### my observation : Increasing of the kernel size make the image more blurry

```python
img = cv2.imread('sharpen.png')

img=cv2.imread("sharpen.png")

img1=cv2.blur(img,(10,10))
img2=cv2.blur(img,(20,20))
img3=cv2.blur(img,(30,30))

f,(ax0,ax1,ax2,ax3)=plt.subplots(1,4,figsize=(30,15))
ax0.imshow(img)
ax1.imshow(img1)
ax2.imshow(img2)
ax3.imshow(img3)
```
