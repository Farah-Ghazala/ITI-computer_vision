```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

### Q1 Make Trackbar for the following images to tune the parameters of canny detection.
### my observation: incresing in values of canny, decreasing of lines & edges

```python
img=cv2.imread("solidWhiteCurve.jpg")
img2=cv2.imread("solidYellowCurve2.jpg")

img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)

img=cv2.GaussianBlur(img,(7,7),0)
img2=cv2.GaussianBlur(img2,(7,7),0)

cv2.namedWindow("SolidCurve")
cv2.namedWindow("SolidCurve2")

def nothing(x):
    pass

cv2.createTrackbar("MinValue","SolidCurve",0,600,nothing)
cv2.createTrackbar("MaxValue","SolidCurve",0,600,nothing)

cv2.createTrackbar("MinValue2","SolidCurve2",0,600,nothing)
cv2.createTrackbar("MaxValue2","SolidCurve2",0,600,nothing)


while True:
    
    MinValue=cv2.getTrackbarPos("MinValue","SolidCurve")
    MaxValue=cv2.getTrackbarPos("MaxValue","SolidCurve")

    MinValue2=cv2.getTrackbarPos("MinValue2","SolidCurve2")
    MaxValue2=cv2.getTrackbarPos("MaxValue2","SolidCurve2")


    edges=cv2.Canny(img,MinValue,MaxValue)
    edges2=cv2.Canny(img2,MinValue2,MaxValue2)
    
    cv2.imshow("SolidCurve",edges)
    cv2.imshow("SolidCurve2",edges2)

    k=cv2.waitKey(1)
    if k==ord('q'):
        break

cv2.destroyAllWindows()
```

### Q2 The out of question 1 make it as input for question 2 after make region of interest then make trackbar for houghlines parameters (threshold ,min_line_lengthmax_line_gap) and then draw 2 line on image.

```python
edges=cv2.Canny(img,100,150)
edges2=cv2.Canny(img2,100,10)


height, width = edges.shape[:2]
vertices = np.array([[(0, height), (width/2, height/2), (width, height)]], dtype=np.int32)
mask = np.zeros_like(edges)
cv2.fillPoly(mask, vertices, 255)
masked_edges = cv2.bitwise_and(edges, mask)

height, width = edges2.shape[:2]
vertices2 = np.array([[(0, height), (width/2, height/2), (width, height)]], dtype=np.int32)
mask2 = np.zeros_like(edges2)
cv2.fillPoly(mask2, vertices2, 255)
masked_edges2 = cv2.bitwise_and(edges2, mask2)

cv2.namedWindow("HOUGH")
cv2.namedWindow("HOUGH2")

def nothing(x):
    pass

cv2.createTrackbar("MIN_VOTES","HOUGH",0,200,nothing)
cv2.createTrackbar("MIN_LINE_LEN","HOUGH",0,100,nothing)
cv2.createTrackbar("MAX_LINE_GAP","HOUGH",0,500,nothing)

cv2.createTrackbar("MIN_VOTES2","HOUGH2",0,200,nothing)
cv2.createTrackbar("MIN_LINE_LEN2","HOUGH2",0,100,nothing)
cv2.createTrackbar("MAX_LINE_GAP2","HOUGH2",0,500,nothing)

while True:
    MIN_VOTES=cv2.getTrackbarPos("MIN_VOTES","HOUGH")
    MIN_LINE_LEN=cv2.getTrackbarPos("MIN_LINE_LEN","HOUGH")
    MAX_LINE_GAP=cv2.getTrackbarPos("MAX_LINE_GAP","HOUGH")

    MIN_VOTES2=cv2.getTrackbarPos("MIN_VOTES2","HOUGH2")
    MIN_LINE_LEN2=cv2.getTrackbarPos("MIN_LINE_LEN2","HOUGH2")
    MAX_LINE_GAP2=cv2.getTrackbarPos("MAX_LINE_GAP2","HOUGH2")


    
    image=cv2.cvtColor(img,cv2.COLOR_GRAY2BGR)
    image2=cv2.cvtColor(img2,cv2.COLOR_GRAY2BGR)

    lines = cv2.HoughLinesP(masked_edges, 1 , np.pi/180 , MIN_VOTES, minLineLength=MIN_LINE_LEN, maxLineGap=MAX_LINE_GAP)
    lines2 = cv2.HoughLinesP(masked_edges2, 1 , np.pi/180 , MIN_VOTES2, minLineLength=MIN_LINE_LEN2, maxLineGap=MAX_LINE_GAP2)

    lines_image = np.zeros((masked_edges.shape[0], masked_edges.shape[1], 3), dtype=np.uint8)
    for line in lines:
        for x1, y1, x2, y2 in line:
            cv2.line(lines_image, (x1, y1), (x2, y2), (0, 0, 255), 5)
    final_image = cv2.addWeighted(image, 0.8, lines_image, 1, 0) 


    lines_image2 = np.zeros((masked_edges2.shape[0], masked_edges2.shape[1], 3), dtype=np.uint8)
    for line in lines2:
        for x1, y1, x2, y2 in line:
            cv2.line(lines_image2, (x1, y1), (x2, y2), (0, 0, 255), 5)
    final_image2 = cv2.addWeighted(image2, 0.8, lines_image2, 1, 0) 
    
    cv2.imshow("HOUGH",final_image)
    cv2.imshow("HOUGH2",final_image2)

    k=cv2.waitKey(1)
    if k==ord('q'):
        break

cv2.destroyAllWindows()
```

### read the following video and apply lane line detection on it then show output using opencv.


```python
cap = cv2.VideoCapture("solidWhiteRight.mp4")

while cap.isOpened():
    state, frame = cap.read()
    if not state:
        cap=cv2.VideoCapture("solidWhiteRight.mp4")
        continue

    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)
    
    
    edges = cv2.Canny(blurred, 50, 150)
    
    
    height, width = frame.shape[:2]
    vertices = np.array([[(0, height), (width/2, height/2), (width, height)]], dtype=np.int32)
    mask = np.zeros_like(edges)
    cv2.fillPoly(mask, vertices, 255)
    masked_edges = cv2.bitwise_and(edges, mask)
    
    
    lines = cv2.HoughLinesP(masked_edges,1, np.pi/180, threshold=50, minLineLength=200, maxLineGap=200)
    
    
    lines_image = np.zeros((masked_edges.shape[0], masked_edges.shape[1], 3), dtype=np.uint8)
    for line in lines:
        for x1, y1, x2, y2 in line:
            cv2.line(lines_image, (x1, y1), (x2, y2), (0, 0, 255), 5)
    
    
    final_image = cv2.addWeighted(frame, 0.8, lines_image, 1, 0)
    
    cv2.imshow('Lane Detection', final_image)
    cv2.imshow('lines_image', lines_image)
    cv2.imshow(' masked_edges', masked_edges)

    k=cv2.waitKey(1)
    if k == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
